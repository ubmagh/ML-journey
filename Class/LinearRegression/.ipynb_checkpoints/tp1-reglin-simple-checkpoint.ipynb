```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from random import seed 
from random import random
```

```python
#generate data
def generatedata(n):
    L = []
    seed(1)
    for i in range(n):
        x = i
        y = 2*x+1+random()
        L.append([x,y])
    L=np.array(L)
    return L
```

```python
X = generatedata(20)
```

```python
for i in range(20):
    plt.scatter(X[i][0],X[i][1])
```

```python
x=[]
y=[]
for i in X:
    x.append(i[0])
    y.append(i[1])
    
```

```python
print(x)
print(y)
```

```python
plt.plot(x,y,"r*")
```

```python
def fit(x,y):
    b1 = np.cov(x,y)[0,1] / np.var(x)
    b0 = np.mean(y) - b1 * np.mean(x)
    return b0,b1
```

```python
b0,b1=fit(x,y)
```

```python
def predict(x,a,b):
    return a*x+b
```

```python
x = np.array(x)
z = predict(x,b1,b0)
```
