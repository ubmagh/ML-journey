## (4) code générale :




```python
def regresseur( W, X): # regression; X here is supposed to receive an additional 1 column on the left side
    return X.dot(W)    
```

```python
def sigmoid( fx ): # for a vector & a number
    sig = lambda x:( 1 / ( 1 + math.exp(-x) ) )
    if type(fx)==type(np.array([])): 
        vectorizedFunc = np.vectorize(sig)
        return vectorizedFunc( fx )
    return sig(fx)
```

```python
sigmoid(3)
```

```python
def func_cout( X, Y, W):
    m = len(Y)
    H0 = sigmoid(regresseur( W, X))
    logH0 = np.log(H0)
    log1_H0 = np.log( 1-H0 )
    return (-1/m)*( Y.dot(logH0)+ (1-Y).dot(log1_H0) ).sum()
```

```python
def gradient( X, Y, W):
    Yc = (sigmoid( regresseur(W, X) ) - Y).dot(X)
    return Yc
```

```python
# tesing
mX = Xand[:, :2]
mX = np.hstack( (np.ones( ( len(mX), 1)), mX ))  
mY = Xand[:, -1]
mW = np.array([ 1, 1, 1])
gradient( mX, mY, mW)

## Pass : manually tested, use https://www.geogebra.org/calculator
```

```python
def MiseJour( grad, W, alpha) :
    return (W-alpha*grad)
```

```python
def Batch_Gradient_Descent(X,y,W,N_max,alpha,eps):
    W_i1 = W
    W_i = W
    
    data = []
    
    for i in range(N_max):
        
        grad = gradient( X, y, W_i)
        W_i1 = MiseJour( grad, W_i, alpha)
        if( abs( func_cout( X, y, W_i ) - func_cout( X, y, W_i1 ) ) <= eps ):
            print("\n➡❗❌Out\n")
            print(  func_cout( X, y, W_i ) , func_cout( X, y, W_i1  ))
            break
        W_i = W_i1 
        data.append( { "w": W_i1, "erreur": func_cout( X, y, W_i1 ) } ) 
    return W_i1, data
```

```python
N_max = 10000
alpha = 0.1
eps = 0.000000001

x = Xand[:, :2]
xx = np.hstack( (np.ones( ( len(x), 1)), x ))    
y = Xand[:, -1]
W = np.array([ 1, 1, 1])

best_w, data = Batch_Gradient_Descent( xx,y,W,N_max,alpha,eps)
print(best_w)
```

```python
def Model_classification( best_W, X):
    get_proba = lambda x : 1 if x>0.5 else 0
    probas = sigmoid( regresseur( best_w, X) )
    if type(X)==type(np.array([])): 
        vectorized_get_proba = np.vectorize(get_proba)
        return vectorized_get_proba(probas)
    return get_proba(probas)
```

```python
Model_classification( best_w, np.array([ 1, 1, 1]) )# allways left 1 for w0
```

```python
# testing with current Xand variables
xtest = Xand[:,:2]
xtest = np.hstack( (np.ones( ( len(xtest), 1)), xtest)) 
Model_classification( best_w, xtest ) # +> identical to y [1,0,0,0]
```

```python
def separateur(W, X1): ## Faire attention à l'ordre sur X: (x1, x2, 1) et W(w2, w1, w0)
    return ( -1/W[2] ) * (W[0]+W[1]*X1)
```
